# 1. Spark MLlib Classification on the Iris Dataset

This notebook demonstrates how to perform a classification task using Apache Spark MLlib.
The Iris dataset is used as the sample dataset. The workflow includes:

- Uploading the Iris dataset
- Loading the dataset into a Spark DataFrame
- Data preprocessing
- Feature vector assembly
- Label encoding
- Splitting data into training and testing sets
- Training a classifier using Spark MLlib
- Evaluating model performance

# 3. Start Spark Session

```python
spark = SparkSession.builder \
    .appName("IrisClassification") \
    .master("local[*]") \
    .getOrCreate()
```

# 4. Upload the File "iris.csv"

```python
# Define schema
schema = StructType([
    StructField("sepal_length", FloatType(), True),
    StructField("sepal_width", FloatType(), True),
    StructField("petal_length", FloatType(), True),
    StructField("petal_width", FloatType(), True),
    StructField("species", StringType(), True)
])

# Load CSV
df = spark.read.csv("iris.csv", schema=schema, header=False)

df.show(5)
df.printSchema()
```
